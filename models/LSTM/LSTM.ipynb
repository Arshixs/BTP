```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, LSTM, Bidirectional, Dense, Dropout,
    BatchNormalization, Concatenate
)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import (
    mean_absolute_error, mean_squared_error, r2_score,
    mean_absolute_percentage_error, median_absolute_error, max_error
)

print(f"TensorFlow version: {tf.__version__}")

TARGET   = 'Mandi Modal Price (AgMarknet)'
SEQUENCE = 150   # How many past days the model looks at to predict the next day
                # Try 60 if your series is long (>500 rows)

# ── 1. Load & Prepare ─────────────────────────────────────────────────────────
print("\nLoading data...")
df = pd.read_excel('../../train.ods', engine='odf')
df['Calendar'] = pd.to_datetime(df['Calendar'])
df = df.sort_values('Calendar').reset_index(drop=True)

# Interpolate any missing dates
df = df.set_index('Calendar').asfreq('D')

# Interpolate only numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].interpolate()

# Fill string columns (optional but recommended)
df[['State', 'Crop', 'Crop_Year']] = df[['State', 'Crop', 'Crop_Year']].ffill()

df = df.reset_index()
print(f"Rows        : {len(df)}")
print(f"Date range  : {df['Calendar'].min().date()} → {df['Calendar'].max().date()}")

# ── 2. Feature Set ────────────────────────────────────────────────────────────
# Price-based features (same logic as XGBoost model)
df['price_lag_1']  = df[TARGET].shift(1)
df['price_lag_7']  = df[TARGET].shift(7)
df['price_lag_30'] = df[TARGET].shift(30)
df['rolling_mean_7']  = df[TARGET].shift(1).rolling(7).mean()
df['rolling_mean_30'] = df[TARGET].shift(1).rolling(30).mean()
df['rolling_std_7']   = df[TARGET].shift(1).rolling(7).std()
df['price_change_1d'] = df[TARGET].shift(1) - df[TARGET].shift(2)
df['price_change_7d'] = df[TARGET].shift(1) - df[TARGET].shift(7)

# Calendar features
df['Month']      = df['Calendar'].dt.month
df['DayOfWeek']  = df['Calendar'].dt.dayofweek
df['DayOfYear']  = df['Calendar'].dt.dayofyear
df['Quarter']    = df['Calendar'].dt.quarter

df = df.dropna().reset_index(drop=True)

PRICE_FEATURES = [TARGET, 'price_lag_1', 'price_lag_7', 'price_lag_30',
                  'rolling_mean_7', 'rolling_mean_30', 'rolling_std_7',
                  'price_change_1d', 'price_change_7d']

EXTERNAL_FEATURES = ['Crude_Price_USD', 'WPI_Paddy', 'Yield', 'MSP_Fixed',
                     'Month', 'DayOfWeek', 'DayOfYear', 'Quarter']

print(f"\nPrice features   : {len(PRICE_FEATURES)}")
print(f"External features: {len(EXTERNAL_FEATURES)}")
print(f"Sequence length  : {SEQUENCE} days")

# ── 3. Scale Features ─────────────────────────────────────────────────────────
# LSTM is sensitive to scale — always normalise to [0, 1]
price_scaler    = MinMaxScaler()
external_scaler = MinMaxScaler()
target_scaler   = MinMaxScaler()

price_scaled    = price_scaler.fit_transform(df[PRICE_FEATURES])
external_scaled = external_scaler.fit_transform(df[EXTERNAL_FEATURES])
target_scaled   = target_scaler.fit_transform(df[[TARGET]])

# ── 4. Build Sequences ────────────────────────────────────────────────────────
# LSTM expects shape: (samples, timesteps, features)
# For each prediction point, we look back SEQUENCE days
def build_sequences(price_data, external_data, target_data, seq_len):
    X_price, X_ext, y = [], [], []
    for i in range(seq_len, len(price_data)):
        X_price.append(price_data[i - seq_len : i])       # (seq_len, price_features)
        X_ext.append(external_data[i])                    # (external_features,) — current day
        y.append(target_data[i])                          # scalar target
    return np.array(X_price), np.array(X_ext), np.array(y)

X_price, X_ext, y = build_sequences(
    price_scaled, external_scaled, target_scaled, SEQUENCE
)
print(f"\nSequence shapes:")
print(f"  X_price : {X_price.shape}   (samples, timesteps, price_features)")
print(f"  X_ext   : {X_ext.shape}     (samples, external_features)")
print(f"  y       : {y.shape}")

# ── 5. Chronological Train / Validation Split ─────────────────────────────────
split = int(len(X_price) * 0.8)
X_price_train, X_price_val = X_price[:split], X_price[split:]
X_ext_train,   X_ext_val   = X_ext[:split],   X_ext[split:]
y_train,        y_val       = y[:split],        y[split:]
print(f"\nTrain: {len(y_train)} | Val: {len(y_val)}")

# ── 6. Model Architecture ─────────────────────────────────────────────────────
#
#   Sequential (price) branch  →  BiLSTM → LSTM → Dropout
#                                                        ↘
#                                                    Concatenate → Dense → Dense → Output
#                                                        ↗
#   External features branch   →  Dense (context encoder)
#
def build_bilstm_model(seq_len, n_price_feat, n_ext_feat):

    # ── Branch 1: Sequential price history ───────────────────────────────────
    price_input = Input(shape=(seq_len, n_price_feat), name='price_input')

    x = Bidirectional(LSTM(128, return_sequences=True), name='bilstm_1')(price_input)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    x = Bidirectional(LSTM(64, return_sequences=True), name='bilstm_2')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.2)(x)

    x = LSTM(32, return_sequences=False, name='lstm_final')(x)
    x = Dropout(0.2)(x)

    # ── Branch 2: External / static features ─────────────────────────────────
    ext_input = Input(shape=(n_ext_feat,), name='ext_input')
    e = Dense(32, activation='relu', name='ext_encoder_1')(ext_input)
    e = BatchNormalization()(e)
    e = Dense(16, activation='relu', name='ext_encoder_2')(e)

    # ── Merge & predict ───────────────────────────────────────────────────────
    merged = Concatenate(name='merge')([x, e])
    out    = Dense(64, activation='relu')(merged)
    out    = Dropout(0.2)(out)
    out    = Dense(32, activation='relu')(out)
    out    = Dense(1,  activation='linear', name='price_output')(out)

    model = Model(inputs=[price_input, ext_input], outputs=out)
    return model

model = build_bilstm_model(SEQUENCE, len(PRICE_FEATURES), len(EXTERNAL_FEATURES))
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='huber',          # More robust to price spikes than MSE
    metrics=['mae']
)
model.summary()

# ── 7. Callbacks ──────────────────────────────────────────────────────────────
callbacks = [
    EarlyStopping(
        monitor='val_loss', patience=20,
        restore_best_weights=True, verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss', factor=0.5,
        patience=8, min_lr=1e-6, verbose=1
    ),
    ModelCheckpoint(
        'best_bilstm.keras', monitor='val_loss',
        save_best_only=True, verbose=0
    )
]

# ── 8. Train ──────────────────────────────────────────────────────────────────
print("\nTraining BiLSTM model...")
history = model.fit(
    [X_price_train, X_ext_train], y_train,
    validation_data=([X_price_val, X_ext_val], y_val),
    epochs=200,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)

# ── 9. Plot Training Curves ───────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
axes[0].plot(history.history['loss'],     label='Train Loss')
axes[0].plot(history.history['val_loss'], label='Val Loss')
axes[0].set_title('Loss (Huber)')
axes[0].legend()
axes[1].plot(history.history['mae'],     label='Train MAE')
axes[1].plot(history.history['val_mae'], label='Val MAE')
axes[1].set_title('MAE')
axes[1].legend()
plt.tight_layout()
plt.savefig('bilstm_training.png', dpi=120)
plt.show()

# ── 10. Metrics ───────────────────────────────────────────────────────────────
y_pred_scaled = model.predict([X_price_val, X_ext_val], verbose=0)
y_pred        = target_scaler.inverse_transform(y_pred_scaled).flatten()
y_actual      = target_scaler.inverse_transform(y_val).flatten()

mae   = mean_absolute_error(y_actual, y_pred)
rmse  = np.sqrt(mean_squared_error(y_actual, y_pred))
mape  = mean_absolute_percentage_error(y_actual, y_pred) * 100
medae = median_absolute_error(y_actual, y_pred)
max_e = max_error(y_actual, y_pred)
r2    = r2_score(y_actual, y_pred)

print("\n── BiLSTM METRICS (Validation Set) ──────────────────────────")
print(f"MAE (Mean Absolute Error):        ₹{mae:.2f}")
print(f"MedAE (Median Absolute Error):    ₹{medae:.2f}")
print(f"Max Error (Worst Prediction):     ₹{max_e:.2f}")
print(f"RMSE:                             ₹{rmse:.2f}")
print(f"MAPE:                             {mape:.2f}%")
print(f"R-squared:                        {r2:.4f}")

# ── 11. Actual vs Predicted Plot ──────────────────────────────────────────────
val_dates = df['Calendar'].iloc[split + SEQUENCE:].reset_index(drop=True)
plt.figure(figsize=(14, 5))
plt.plot(val_dates, y_actual, label='Actual',    color='steelblue', linewidth=1.5)
plt.plot(val_dates, y_pred,   label='Predicted', color='crimson',
         linewidth=1.5, linestyle='--')
plt.title('BiLSTM — Mandi Price: Actual vs Predicted')
plt.ylabel('Price (₹)')
plt.legend()
plt.tight_layout()
plt.savefig('bilstm_forecast.png', dpi=120)
plt.show()

# ── 12. Submission Function ───────────────────────────────────────────────────
def generate_lstm_submission(test_filepath, trained_model,
                             full_df, p_scaler, e_scaler, t_scaler):
    """
    Iterative prediction: feeds each predicted price back as lag input
    for the next step — same strategy as the XGBoost submission function.
    """
    print(f"\nProcessing test file: {test_filepath}")

    df_test = pd.read_excel(test_filepath, engine='odf')
    df_test['Calendar'] = pd.to_datetime(df_test['Calendar'])
    df_test = df_test.sort_values('Calendar').reset_index(drop=True)

    last_known = full_df.iloc[-1]
    for col in ['Crude_Price_USD', 'WPI_Paddy', 'Yield', 'MSP_Fixed']:
        if col not in df_test.columns or df_test[col].isna().all():
            df_test[col] = last_known[col]

    # Seed the rolling buffer with last SEQUENCE rows of real data
    buffer = full_df[PRICE_FEATURES].tail(SEQUENCE).values.copy()  # (SEQUENCE, features)
    predicted_prices = []

    for _, row in df_test.iterrows():
        # Scale the current buffer and predict
        buf_scaled = p_scaler.transform(buffer)                          # (SEQUENCE, price_feat)
        ext_row    = pd.DataFrame([{
            'Crude_Price_USD': row['Crude_Price_USD'],
            'WPI_Paddy'      : row['WPI_Paddy'],
            'Yield'          : row['Yield'],
            'MSP_Fixed'      : row['MSP_Fixed'],
            'Month'          : row['Calendar'].month,
            'DayOfWeek'      : row['Calendar'].dayofweek,
            'DayOfYear'      : row['Calendar'].dayofyear,
            'Quarter'        : row['Calendar'].quarter,
        }])
        ext_scaled = e_scaler.transform(ext_row)

        pred_scaled = trained_model.predict(
            [buf_scaled[np.newaxis], ext_scaled], verbose=0
        )
        pred_price = t_scaler.inverse_transform(pred_scaled)[0, 0]
        predicted_prices.append(pred_price)

        # Build a new row for the price buffer using the predicted price
        new_price_row = np.array([[
            pred_price,
            buffer[-1, 0],           # lag_1  = previous prediction
            buffer[-6, 0],           # lag_7
            buffer[-29, 0] if len(buffer) >= 30 else buffer[0, 0],  # lag_30
            np.mean(buffer[-7:, 0]), # rolling_mean_7
            np.mean(buffer[-30:, 0]) if len(buffer) >= 30 else np.mean(buffer[:, 0]),
            np.std(buffer[-7:, 0]),  # rolling_std_7
            buffer[-1, 0] - buffer[-2, 0] if len(buffer) >= 2 else 0,  # change_1d
            buffer[-1, 0] - buffer[-7, 0] if len(buffer) >= 7 else 0,  # change_7d
        ]])

        # Slide the window: drop oldest row, append new
        buffer = np.vstack([buffer[1:], new_price_row])

    submission = pd.DataFrame({
        'Calendar'       : df_test['Calendar'],
        'Predicted_Price': predicted_prices
    })
    submission.to_csv('submission_bilstm.csv', index=False)

    print(f"✓ {len(submission)} predictions → submission_bilstm.csv")
    print(f"  Price range : ₹{min(predicted_prices):.2f} – ₹{max(predicted_prices):.2f}")
    print(f"  Mean price  : ₹{np.mean(predicted_prices):.2f}")
    return submission


```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import (
    mean_absolute_error, 
    mean_squared_error, 
    r2_score,
    mean_absolute_percentage_error,
    median_absolute_error,
    max_error,
    explained_variance_score
)

print("Loading test results from submission.csv...")

# 1. Load the updated submission file
df_results = pd.read_csv('submission_bilstm.csv')

# Saftey check: Drop any rows where the Actual_Price might be accidentally blank
df_results = df_results.dropna(subset=['Actual_Price', 'Predicted_Price'])

# 2. Extract the actual and predicted columns
y_actual = df_results['Actual_Price']
y_pred = df_results['Predicted_Price']

# 3. Calculate all the metrics
mae = mean_absolute_error(y_actual, y_pred)
rmse = np.sqrt(mean_squared_error(y_actual, y_pred))
mape = mean_absolute_percentage_error(y_actual, y_pred) * 100  # Convert to percentage
medae = median_absolute_error(y_actual, y_pred)
max_err = max_error(y_actual, y_pred)
r2 = r2_score(y_actual, y_pred)
evs = explained_variance_score(y_actual, y_pred)

# 4. Print the final scorecard
print("\n--- FINAL TEST SET METRICS (submission.csv) ---")
print(f"MAE (Mean Absolute Error):        ₹{mae:.2f}")
print(f"MedAE (Median Absolute Error):    ₹{medae:.2f}")
print(f"Max Error (Worst Prediction):     ₹{max_err:.2f}")
print('-' * 48)
print(f"RMSE (Root Mean Squared Error):   ₹{rmse:.2f}")
print(f"MAPE (Mean Abs Percentage Error): {mape:.2f}%")
print('-' * 48)
print(f"R-squared:                        {r2:.4f}")
print(f"Explained Variance Score:         {evs:.4f}")
```
