```python
import pandas as pd

# ---------------------------------------------------------
# 1. Opening the CSV files (.csv)
# ---------------------------------------------------------
# CSVs are the most straightforward format to read natively in pandas.
df_up_paddy = pd.read_csv('Uttar-Pradesh,-Paddy,-(Jan-01,-2011-to-Apr-02,-2026).csv')
df_usd_inr = pd.read_csv('USD_INR Historical Data.csv')


print("Loaded CSV files successfully.")


# ---------------------------------------------------------
# 3. Opening the OpenDocument Spreadsheets (.ods)
# ---------------------------------------------------------
# We use read_excel again, but specify the 'odf' engine.
df_wpi_paddy = pd.read_excel('WPI_Paddy.ods', engine='odf')
df_msp = pd.read_excel('MSP.ods', engine='odf')

print("Loaded ODS files successfully.")

# ---------------------------------------------------------
# Quick Check: View your data
# ---------------------------------------------------------
# Use .head() to print the first 5 rows of any DataFrame to ensure it loaded correctly.
print("\nPreview of Uttar Pradesh Paddy Data:")
print(df_up_paddy.head())
```

```python
import pandas as pd

# 1. Convert both date columns to standard datetime formats so they match perfectly
df_up_paddy['Calendar'] = pd.to_datetime(df_up_paddy['Calendar'])
df_usd_inr['Date'] = pd.to_datetime(df_usd_inr['Date'], format='%d-%m-%Y')

# 2. Extract only the columns you need from the USD_INR dataframe (Date and Price)
# It's a good practice to rename 'Price' so you know what it is in the final table
df_usd_subset = df_usd_inr[['Date', 'Price']].rename(columns={'Price': 'USD_INR_Price'})

# 3. Merge (Append) the datasets
# We use a 'left' join to keep all original rows in df_up_paddy and only bring over matching prices
df_final = pd.merge(
    df_up_paddy, 
    df_usd_subset, 
    left_on='Calendar',   # The date column in your left table (Paddy)
    right_on='Date',      # The date column in your right table (USD/INR)
    how='left'            # 'left' ensures we don't lose any Paddy data if a date is missing in USD/INR
)

# 4. Clean up: Drop the redundant 'Date' column that was brought over from df_usd_subset
df_final = df_final.drop(columns=['Date'])

# View the result
print("Merged DataFrame:")
print(df_final.head())

# Fills NaN values with the previous valid day's price
df_final['USD_INR_Price'] = df_final['USD_INR_Price'].ffill()

# If the very FIRST row is a weekend, ffill won't catch it. 
# We can use bfill (backward fill) just to be safe for the start of the dataset.
df_final['USD_INR_Price'] = df_final['USD_INR_Price'].bfill()

print("Missing values filled!")
```

```python
import pandas as pd

df = pd.read_excel("APY.ods", engine="odf")

# Forward-fill merged State and District columns
df['State']    = df['State'].ffill()
df['District'] = df['District'].ffill()

df_up_aggregated = df.groupby(['State', 'Year'])[['Area (Hectare)', 'Production (Tonnes)']].sum().reset_index()
df_up_aggregated = df_up_aggregated.rename(columns={'Area (Hectare)': 'Area','Production (Tonnes)':"Production"})
df_up_aggregated['Yield'] = (df_up_aggregated['Production'] / df_up_aggregated['Area']).round(2)
```

```python
import pandas as pd

# 1. Define a function to map a specific date to the Indian Kharif Crop Year
def get_crop_year(date):
    # If the month is October (10) or later, it belongs to the current year's harvest
    if date.month >= 10:
        return f"{date.year} - {date.year + 1}"
    # If the month is before October, it is still trading the previous year's harvest
    else:
        return f"{date.year - 1} - {date.year}"

# 2. Apply this function to your daily price dataframe to create a joining key
# Assuming your date column in df_final is a standard pandas datetime object
df_final['Crop_Year'] = df_final['Calendar'].apply(get_crop_year)

# 3. Merge the datasets!
# We use a 'left' join to keep all daily prices, bringing in the Yield data where the Crop_Years match
df_model_ready = pd.merge(
    df_final,
    df_up_aggregated[['Year', 'Area', 'Production', 'Yield']], # Only bring the columns you need
    left_on='Crop_Year',
    right_on='Year',
    how='left'
)

# 4. Clean up by dropping the redundant 'Year' column from the right table
df_model_ready = df_model_ready.drop(columns=['Year'])

print(df_model_ready[['Calendar', 'Mandi Modal Price (AgMarknet)', 'Crop_Year', 'Yield']].head(10))
```

```python
import pandas as pd

# 1. Read your simplified MSP file
# We only take the first 3 columns to ignore any blank trailing columns
df_msp_new = pd.read_excel('MSP.ods', engine='odf').iloc[:, 0:3]

# Rename the columns to make them easy to work with
df_msp_new.columns = ['Raw_Year', 'MSP_Rec', 'MSP_Fixed']

# 2. Clean the data
# Drop any rows where the price is just a dash ('-') or empty
df_msp_new = df_msp_new[df_msp_new['MSP_Fixed'] != '-']

# Convert the Fixed Price column to actual numbers
df_msp_new['MSP_Fixed'] = pd.to_numeric(df_msp_new['MSP_Fixed'], errors='coerce')

# 3. Standardize the Year Format
# We need "2024-25" to become "2024 - 2025" to match our Yield data
def format_msp_year(y):
    if pd.isna(y): return y
    parts = str(y).strip().split('-')
    if len(parts) == 2:
        # Add "20" to the second part (e.g., '25' becomes '2025')
        end_year = f"20{parts[1]}" if len(parts[1]) == 2 else parts[1]
        return f"{parts[0]} - {end_year}"
    return y

df_msp_new['Crop_Year'] = df_msp_new['Raw_Year'].apply(format_msp_year)

# 4. Merge it into your main dataset!
# Assuming 'df_model_ready' is the dataframe from our previous step
df_final_model = pd.merge(
    df_model_ready, 
    df_msp_new[['Crop_Year', 'MSP_Fixed']], # Only bring over the formatted year and the fixed price
    on='Crop_Year',
    how='left'
)

print("Data successfully merged!")
print(df_final_model[['Calendar', 'Mandi Modal Price (AgMarknet)', 'Yield', 'MSP_Fixed']].head(10))
```

```python
import pandas as pd

# 1. Read the simplified WPI file
# We grab the first 3 columns and set the first row as the header
df_wpi_new = pd.read_excel('WPI_Paddy_1.ods', engine='odf', header=0).iloc[:, 0:3]

# 2. Clean out the missing data
# Remove the rows from early 2012 where WPI is just a dash ('-')
df_wpi_new = df_wpi_new[df_wpi_new['WPI_Paddy'] != '-']

# Convert WPI_Paddy to a proper numeric type
df_wpi_new['WPI_Paddy'] = pd.to_numeric(df_wpi_new['WPI_Paddy'], errors='coerce')

# 3. Create the 'Year_Month' merging key
# Map the 3-letter month names to 2-digit numbers
month_map = {
    'Jan': '01', 'Feb': '02', 'Mar': '03', 'Apr': '04', 
    'May': '05', 'Jun': '06', 'Jul': '07', 'Aug': '08', 
    'Sep': '09', 'Oct': '10', 'Nov': '11', 'Dec': '12'
}

# Apply the mapping to create a "YYYY-MM" string (e.g., "2012-04")
df_wpi_new['Month_Num'] = df_wpi_new['Month'].map(month_map)
df_wpi_new['Year_Month'] = df_wpi_new['Year'].astype(str) + '-' + df_wpi_new['Month_Num']

# Keep only the columns we need for the merge
df_wpi_clean = df_wpi_new[['Year_Month', 'WPI_Paddy']]

# ------------------------------------------------------------------
# 4. Merge into your Main Model Dataset
# ------------------------------------------------------------------

# Create the same 'Year_Month' key in your main dataset
# Assuming 'Calendar' is already a datetime object
df_final_model['Year_Month'] = df_final_model['Calendar'].dt.strftime('%Y-%m')

# Perform the merge
df_final_model = pd.merge(
    df_final_model,
    df_wpi_clean,
    on='Year_Month',
    how='left'
)

# 5. Clean up the final dataset
# Drop the temporary joining column
df_final_model = df_final_model.drop(columns=['Year_Month'])

# Safely drop any rows from before April 2012 (where WPI is NaN because we didn't have data)
df_final_model = df_final_model.dropna(subset=['WPI_Paddy'])

print("WPI Data Successfully Merged!")
print(df_final_model[['Calendar', 'Mandi Modal Price (AgMarknet)', 'Yield', 'MSP_Fixed', 'WPI_Paddy']].head())
```

```python
import pandas as pd
import yfinance as yf

print("Fetching Brent Crude Oil data...")

# 1. Download historical Crude Oil data (Brent Crude)
# We set the start date to match the beginning of your truncated dataset
crude_data = yf.download('BZ=F', start='2012-04-01', end='2026-04-02')

# 2. Clean and format the fetched data
# yfinance puts the Date in the index. We need to reset it to a regular column.
crude_data = crude_data.reset_index()

# We only care about the 'Date' and the 'Close' price
crude_clean = crude_data[['Date', 'Close']].copy()

# Rename the columns to be clear
# 'Close' can sometimes be a MultiIndex depending on yfinance versions, 
# so we force the column names to be flat strings.
crude_clean.columns = ['Calendar', 'Crude_Price_USD']

# Convert the Date to standard pandas datetime (just to be absolutely safe)
crude_clean['Calendar'] = pd.to_datetime(crude_clean['Calendar']).dt.tz_localize(None)

# 3. Merge into your Main Model Dataset
# Assuming your date column in df_final_model is called 'Calendar'
df_final_model['Calendar'] = pd.to_datetime(df_final_model['Calendar']).dt.tz_localize(None)

df_final_model = pd.merge(
    df_final_model,
    crude_clean,
    on='Calendar',
    how='left'
)

# 4. Handle Forex/Commodity Weekends
# Crude oil doesn't trade on weekends. We forward-fill the Friday prices into Saturday and Sunday.
df_final_model['Crude_Price_USD'] = df_final_model['Crude_Price_USD'].ffill()

# If the very first day happens to be a weekend, back-fill just to catch it
df_final_model['Crude_Price_USD'] = df_final_model['Crude_Price_USD'].bfill()

print("Crude Oil Data Successfully Merged!")

# Let's see the glorious final dataset
print(df_final_model[['Calendar', 'Mandi Modal Price (AgMarknet)', 'Crude_Price_USD', 'WPI_Paddy', 'Yield', 'MSP_Fixed']].head(10))

# Save your final, ML-ready dataset to a CSV
df_final_model.to_csv('Master_Agri_Dataset.csv', index=False)
print("Pipeline complete! Saved to Master_Agri_Dataset.csv")
```
